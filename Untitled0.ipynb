```python
import pandas as pd
import numpy as np  # Make sure numpy is imported
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train=pd.read_csv('/content/drive/MyDrive/train.csv')
```

```python
# Replace '?' with NaN
train.replace('?', np.nan, inplace=True)
```

```python
train.info ()
```

```python
train.head()
```

```python
print(train['Fare'].isnull().sum())  # This should print 0 if all missing values were filled
```

```python
# Convert 'Fare' to numeric (setting non-convertible values to NaN)
train['Fare'] = pd.to_numeric(train['Fare'], errors='coerce')
```

```python
# Impute missing values in 'Fare' with the mean of the column
train['Fare'].fillna(train['Fare'].mean(), inplace=True)
```

```python
# Verify that no missing values remain in 'Fare'
print(f"Missing values in 'Fare': {train['Fare'].isnull().sum()}")
```

```python
# Calculate and print the mean of 'Fare'
mean_fare = train['Fare'].mean()
print(f"Mean of 'Fare': {mean_fare}")
```
